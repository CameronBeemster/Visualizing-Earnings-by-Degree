## Exploration

__Do students in more popular majors make more money?__

Looking at the last two plots _ax5_ and _ax6_, there is no evidence that the students in more popular majors (the majors with a higher number of male/female students) make more money. In fact, based on these graphs, it appears that only the students in the least popular majors have the higher earning potential.

__Do students that majored in subjects that were majority female make more money?__

In plot _ax4_, there seems to be a weak, negative correlation between the share of women in the relative major and median salary. In other words, as the female percentage of a major increases there is a slight decrease in the median salary. The correlation appears to be weak, although there is a noticable trend.

__Is there any link between the number of full-time employees and median salary?__

The distribution of data in plot _ax3_ shows that there is no link between the number of full-time students and median salary. This has to do with the fact that less full-time students may imply a less popular major and a smaller sample size. This can artificially inflate or deflate medians. So, all the data points closer to zero full-time employees with high median salaries could just be due to random chance.

## Exploration

The scatter plots displaying _Full time_, _Part time_, and _Total_ graduates with _Unemployment rate_ gives an insight into the likelihood of employment after graduation. It appears the three of these graphs follow the same pattern - the more students a program has, the more likely that 5-10% gradutes will experience unemployment. 

The unemployment rate between 2010 and 2012 hovered around 10%, which is why the relationship between _Total_ graduates and _Unemployment rate_ trends in the range of 5-10%. It is likely the more graduates of a specifc major there were, the more saturated the market was for those graduates, and thus a 5-10% unemployment makes sense.

The majors with less graduates gives another insight, as well. Programs with a small number of graduates can have anywhere between a ~0% or ~20% unemployment rate. The reason behind this could be that these smaller programs are undesirable because they are very challenging (higher median salary) or very niche (lower median salary).
